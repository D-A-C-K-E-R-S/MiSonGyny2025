# Importaciones

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

# Task1

```python
# Cargar el archivo CSV
df_task1 = pd.read_csv("./data/train/subtask1_train_rev002.csv")  # Reemplaza con el nombre de tu archivo

# Obtener la distribución de etiquetas
label_distribution = df_task1["label"].value_counts()
label_percentage = df_task1["label"].value_counts(normalize=True) * 100

# Crear un DataFrame con los resultados
distribution_df = pd.DataFrame({
    "Label": label_distribution.index,  # Se agrega la columna Label
    "Número de instancias": label_distribution.values,
    "Porcentaje": label_percentage.values
})

# Graficar la distribución con colores distintos
plt.figure(figsize=(8, 5))
colors = sns.color_palette("husl", len(distribution_df))  # Paleta de colores distinta para cada barra
plt.bar(distribution_df["Label"], distribution_df["Número de instancias"], color=colors, edgecolor="black")

# Personalización del gráfico
plt.xlabel("Label")
plt.ylabel("Frecuencia")
plt.title("Distribución de Labels")
plt.xticks(rotation=45)

# Mostrar la gráfica
plt.show()

# Mostrar la tabla
distribution_df.head()
```

```python
# Crear una nueva columna con el texto limpio
df_task1["lyrics_clean"] = df_task1["lyrics"].str.replace(r"\[.*?\]", "", regex=True).str.replace(r"\s+", " ", regex=True).str.strip()

# Verificar cambios
df_task1[["lyrics", "lyrics_clean"]].head()
```

```python
# Estadísticas básicas: calcular la longitud de los textos (en número de palabras)
df_task1["text_length"] = df_task1["lyrics_clean"].apply(lambda x: len(x.split()))

# Histograma de la longitud de los textos en el conjunto de entrenamiento
plt.figure(figsize=(10, 5))
sns.histplot(df_task1["text_length"], bins=50, kde=True, color="blue")
plt.title("Distribución de la longitud de los textos (train)")
plt.xlabel("Número de palabras")
plt.ylabel("Frecuencia")
plt.show()
```
